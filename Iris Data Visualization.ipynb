```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = sns.load_dataset('iris')
print(a)

#                                            *****RELPLOT*****

sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'line')

#LINE PLOT
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species')

#SCATTER PLOT with HUES
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species', style = 'species')

#STYLE changes the hue symbols
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species', style = 'species', size = 'sepal_width')

#Assigning SIZE to a certain column represents the size of values(icons) with in the column's range, in below case its 2 to 4 from sepal_width
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species', style = 'species', size = 'sepal_length', sizes=(10, 100))

#SIZES parameter provides the size of the values, in this case starting from 10 and ending with 100.
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species', style = 'species', size = 'sepal_length', sizes=(10, 100), col = 'species')

#COLUMN divides the presentation into categorical values represented in a column order, 3 species in this case.
```

```python
sns.relplot(data = a, x = 'sepal_length', y = 'sepal_width', kind = 'scatter', hue = 'species', style = 'species', size = 'sepal_length', sizes=(10, 100), row = 'species')

#ROW divides the presentation into categorical values represented in a row order, 3 species in this case.
```

```python
#                                              *****DISPLOTS*****

a = sns.load_dataset('tips')

print(a)

# *UNIVARIATE* Only one variable which is total_bill, with default number of bins.

sns.displot(data = a, x = 'total_bill')
```

```python
# Custom bin value given, as 10.

sns.displot(data = a, x = 'total_bill', bins = 10)
```

```python
#SWITCHING axis

sns.displot(data = a, y = 'total_bill', bins = 10)
```

```python
# Binwidth

sns.displot(data = a, x = 'total_bill', binwidth = 3)
```

```python
#BINRANGE lets us select which range of data to plot, in this case 10 to 40.

sns.displot(data = a, x = 'total_bill', bins = 10, binrange = (10,40))
```

```python
#WITH HUE

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex') 
```

```python
#WITH HUE being stacked

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex', multiple="stack") 
```

```python
#WITH HUE being dodged

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex', multiple="dodge") 
```

```python
#To remove LEGENDS

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex', multiple="dodge", legend = False) 
```

```python
#Element parameter lets us change the type of chart from bar, step and poly. Step was used here

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex', multiple="dodge", legend = False, element = 'step') 
```

```python
# Stat gives count, frequency, probability and density between bins. Frequency used here

sns.displot(data = a, x = 'total_bill', bins = 10, hue = 'sex', multiple="dodge", legend = False, stat = 'frequency' ) 
```

```python
sns.kdeplot(data=a, x="total_bill")
```

```python
#Adjusting bandwith, 0.2 in this case. This is how the presentation is smoothened. By default bw is 1.

sns.kdeplot(data=a, x="total_bill", bw_adjust=0.2)
```

```python
# Multiple 

sns.kdeplot(data=a, x="total_bill", hue="time", multiple="stack")
```

```python
#                                      ****ECDF-PLOT****

sns.ecdfplot(data = a, x = "total_bill")
```

```python
#Stat parameter: “proportion”, “count”

sns.ecdfplot(data=a, x="total_bill", hue="sex", stat="count")
```

```python
#                                       ****RUGPLOT

# This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way.

sns.scatterplot(data=a, x="total_bill", y="tip")
sns.rugplot(data=a, x="total_bill", y="tip")
```

```python
#                                        ***STRIPPLOT***

import seaborn as sns

tips = sns.load_dataset("tips")

sns.stripplot(x=tips["total_bill"])
```

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
#Jitter defines how close the data is packed

sns.stripplot(x="day", y="total_bill", data=tips, jitter=0.05)
```

```python
sns.stripplot(x="sex", y="total_bill", hue="day", data=tips)
```

```python
#Control strip order by passing an explicit order:

sns.stripplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])
```

```python
import numpy as np

sns.boxplot(x="tip", y="day", data=tips, whis=np.inf)
sns.stripplot(x="tip", y="day", data=tips, color=".3")
```

```python
# SWARMPLOT

sns.swarmplot(x=tips["total_bill"])
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
# BOXPLOT

sns.boxplot(x=tips["total_bill"])
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex",
                    data=tips, palette="Set2", split=True,
                    scale="count")
```
